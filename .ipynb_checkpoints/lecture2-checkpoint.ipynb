```python
# Pixel-points with different strength in their color in a range/scale of colour
# eg. from black range to white range
x = [130, 240, 113, 55, 128]

# Set the value to threshold, a specific point of color range/ quality your model needs to optimize
# for eg. from black to white range, choose the threshold of 127

threshold = 127
for i in range(len(x)):
    if x[i] > threshold: 
        x[i] = 255 # whether if the checked point/pixel is of Black
    else:
        x[i] = 0 # whether if the checked point/pixel is of White
print(x)
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

path = "../MachineVision/Gold_coins.jpg"  # Replace with the correct path to your image file
image = cv.imread(path)

if image is not None:
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))  # Convert BGR to RGB for Matplotlib
    plt.title("original")
    plt.show()
else:
    print("Image not found or could not be loaded.")
```

```python
image = image[:,:,::-1]
plt.imshow(image)
plt.title("original")
plt.show()
```

```python
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)  # Corrected the color conversion
plt.imshow(image_gray, cmap='gray')  # Specify the colormap as 'gray'
plt.title("gray")
plt.show()

# Does BRG-order matter? It matters, because in somecase, the picture is distored in someway
# SO, Math can work, if math works, calculation succeeds and the model works
```

```python
plt.imshow(image_gray, cmap="gray")
plt.title("gray")
plt.show()
```

```python
# When makes image to become smaller. we will loose data from the image
# pixel may be broken, and it get blurred, loosing details

image_small = cv.resize(image_gray, (200,160))
plt.imshow(image_small, cmap="gray")
plt.title("gray")
plt.show()
```

```python
# CONVENTION in Python: let the variable whihc are not important as _ 'underscore'

threshold = 127
value = 255
_, image_small_whit_threshold = cv.threshold(image_small, threshold, value, cv.THRESH_BINARY)
plt.imshow(image_small_whit_threshold, cmap="Greys")
plt.title("blackwhite")
plt.show()
```
